```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
'What we are doing is that we are investigating if missing values have a pattern because it can lead to bias in analysis. for example missing values are linked to about the books in the dataset. We are gonna use seaborn and create a missingness heat map'

ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar = False, ax=ax)
plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()

'subtitle is mostly missing as we can see on the heatmap (there are gaps), the categories, description and thumbnails show a random pattern, whereas average_rating, num_pages, and ratings_count it shows that observations that have one of them missing have all of them missing this could be from another dataset. this can lead to bias as some books are just newer and not rated etc'
```

```python
'Description field of the dataset is something I am going to use as a core field for my NLP work. So I want to also observe if there is a pattern for books with missing descriptions. I will probably get rid of these books and I need to know whether the recommender skews to a certain type of books'

import numpy as np
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]

```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]
correlation_matrix = books[columns_of_interest].corr(method = "spearman")

sns.set_theme(style = "white")
plt.figure(figsize = (8,8))
heatmap = sns.heatmap(correlation_matrix, annot = True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()

'Since the correlations are weak, it gives me enough confidence that the recommender will not be skewed to any particular type of books'
```

```python
'Since only 5% of the books have the missing description or missing num_pages field or average rating or published_year field, I decided to discard them'
final_books = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
]

final_books
```

```python
'Some categories are too specific and hence i want to see the distribution'
final_books["categories"].value_counts().reset_index().sort_values("count", ascending = False)
```

```python
'Description field should also makes sense and be useful so I should be checking and cleaning it too'
final_books["words_in_description"] = final_books["description"].str.split().str.len()
final_books
```

```python
final_books_25_words = final_books[final_books["words_in_description"]>=25]
```

```python
'Some subtitles are missing'
final_books_25_words["title_and_subtitle"] = (np.where(final_books_25_words["subtitle"].isna(), final_books_25_words["title"],
                                                       final_books_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis = 1)))
```
